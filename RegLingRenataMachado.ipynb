```python
import pandas as pd # conventional alias
from sklearn.datasets import load_boston

dataset = load_boston()

dt = pd.DataFrame(dataset.data, columns=dataset.feature_names,)

#top 5 de registros da tabela
dt.head()
```

```python
#observacoes e colunas 

dt.shape
```

```python
print("dt.keys(): \n{}".format(dt.keys()))
```

```python
#campo Target(respostas) dessa base
#'PTRATIO'
print(dataset.target)
```

```python
print(dataset.feature_names)
```

```python
dt['PRICE'] = dataset.target
print(dt.head())
```

```python
#biblioteca para plotar dados 
import seaborn as sns

%matplotlib inline
```

```python
#5-Usando a biblioteca de visualização seaborn, plote o gráfico que mostra a relação entre as features e responses
sns.pairplot(dt, x_vars=['CRIM', 'ZN' ,'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS','PTRATIO', 'RAD' ,'TAX','B', 'LSTAT'], y_vars='PRICE', size=5, aspect=0.5, kind='reg')
```

```python
#6-Prepare X e y usando o pandas
feature_cols = [ 'CRIM', 'ZN' ,'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS','PTRATIO', 'RAD' ,'TAX','B', 'LSTAT']

x= dt[feature_cols]

x.head()
```

```python
#6-Prepare X e y usando o pandas
#7-Qual o tipo de dados de X e y?
y= dt['PRICE']

y = dt.PRICE

y.head()
```

```python
#8-Sobre o que se trata essa base de dados? Que tipo de informações ela guarda?
#PREÇOS DE CASAS DE BOSTON
```

```python

from sklearn.cross_validation import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=1)
```

```python
#9-Gere um X de treino e y de treino, X de teste e y da base (Split Train/Test)
#divisao padrão: 75% para treino
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression 

linreg = LinearRegression()

#treinando os dados
linreg.fit(x_train, y_train)
```

```python
print(linreg.intercept_)
print(linreg.coef_)
```

```python
list(zip(feature_cols, linreg.coef_))
```

```python
#gerando a previsao baseado no valor do teste
y_pred = linreg.predict(x_test)
```

```python
from sklearn import metrics
import numpy as np
#gera a previsão da média de erro. a previsão ideal e zero
print (np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
